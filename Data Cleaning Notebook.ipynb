USA datframe cleaned to show year and GDP growth without extra features:

```python
df_USA_2 = df.loc[df['Country Name'] == "United States"]

X_USA = df_USA_2.loc[df_USA_2["Indicator Name"] == "GDP growth (annual %)"]

X_USA = X_USA.drop(columns=["Country Name", "Country Code", "Indicator Name", "Indicator Code", "1960", "2020","2050", "Unnamed: 66"])
X_USA = X_USA.T

X_USA.rename(columns = {15433:"GDP growth (annual %)"}, inplace = True) 
X_USA.rename(columns = {" ":'Date'}, inplace = True) 

X_USA["date"] = X_USA.index

X_USA["date"] = pd.to_datetime(X_USA["date"])


df_gdp_in_dollars_USA = df_gdp_in_dollars[["United States GDP (Dollars)", "date"]].iloc[1:-1]
df_gdp_in_dollars_USA["United States GDP (Dollars)"] = df_gdp_in_dollars_USA["United States GDP (Dollars)"].astype(float) 

date_time_merge_column = []
for i in X_USA["date"]:
    date_time_merge_column.append(i)
df_gdp_in_dollars_USA["date"] = date_time_merge_column

X_USA = pd.merge(X_USA, df_gdp_in_dollars_USA) 
```
